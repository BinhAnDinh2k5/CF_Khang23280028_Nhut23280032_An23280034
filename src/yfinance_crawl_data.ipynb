```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python


ticker_df = pd.read_csv(r'C:\Users\Dinh Binh An\OneDrive\Dai_hoc\toan_tai_chinh\giua_ki\tickers.csv') 
ticker_symbol_list = list(ticker_df['Ticker'])

for ticker_symbol in ticker_symbol_list:
    try:
        ticker = yf.Ticker(ticker_symbol)
    
        # Lấy dữ liệu 10 năm trở về trước
        historical_data = ticker.history(end="2025-10-01", period="10y")
        
        # Kiểm tra số lượng dữ liệu thực tế
        if historical_data.empty:
            print(f"{ticker_symbol} không có dữ liệu, bỏ qua")
            continue
        
        # Tính khoảng thời gian thực tế
        start_date = historical_data.index[0].to_pydatetime()
        end_date = historical_data.index[-1].to_pydatetime()
        years_of_data = (end_date - start_date).days / 365.25  # xấp xỉ năm
        
        if years_of_data < 7:
            print(f"{ticker_symbol} chỉ có {years_of_data:.2f} năm dữ liệu, bỏ qua")
            continue
        
        # Lưu CSV nếu đủ 7 năm trở lên
        historical_data.to_csv(
            f'C:\\Users\\Dinh Binh An\\OneDrive\\Dai_hoc\\toan_tai_chinh\\giua_ki\\Data\\{ticker_symbol}.csv'
        )
        
        print(f"Đã lưu dữ liệu ticker {ticker_symbol} ({years_of_data:.2f} năm)")
        
    except Exception as e:
        print(f"Không crawl được dữ liệu ticker: {ticker_symbol}, lỗi: {e}")
```
